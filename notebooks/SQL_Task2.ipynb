```python
query = """
SELECT *
FROM superstore
LIMIT 5;
"""

pd.read_sql(query, conn)
```

# Business Question 1

Which region generates the highest profit?


```python
query = """
SELECT
    region,
    ROUND(SUM(CAST(profit AS REAL)),2) AS total_profit
FROM superstore
GROUP BY region
ORDER BY total_profit DESC;
"""

pd.read_sql(query, conn)
```

# Business Question 2

Which are the Top 5 Cities by Sales?

```python
query = """
SELECT
    city,
    ROUND(SUM(CAST(sales AS REAL)),2) AS total_sales
FROM superstore
GROUP BY city
ORDER BY total_sales DESC
LIMIT 5;
"""

pd.read_sql(query, conn)
```

# Business Question 3

Which category generates the highest sales?

```python
query = """
SELECT
    category,
    ROUND(SUM(CAST(sales AS REAL)),2) AS total_sales
FROM superstore
GROUP BY category
ORDER BY total_sales DESC;
"""

pd.read_sql(query, conn)
```

# Business Question 4

Which category generates the highest profit?

```python
query = """
SELECT
    category,
    ROUND(SUM(CAST(profit AS REAL)),2) AS total_profit
FROM superstore
GROUP BY category
ORDER BY total_profit DESC;
"""

pd.read_sql(query, conn)
```

# Business Question 5

What are the Top 5 Products (Sub-Categories) by Sales?

```python
query = """
SELECT
    sub_category,
    ROUND(SUM(CAST(sales AS REAL)),2) AS total_sales
FROM superstore
GROUP BY sub_category
ORDER BY total_sales DESC
LIMIT 5;
"""

pd.read_sql(query, conn)
```

# Business Question 6

Which ship mode is used most frequently?

```python
query = """
SELECT
    ship_mode,
    COUNT(*) AS total_orders
FROM superstore
GROUP BY ship_mode
ORDER BY total_orders DESC;
"""

pd.read_sql(query, conn)
```

# Business Question 7

Which customer segment generates the highest sales?

```python
query = """
SELECT
    segment,
    ROUND(SUM(CAST(sales AS REAL)),2) AS total_sales
FROM superstore
GROUP BY segment
ORDER BY total_sales DESC;
"""

pd.read_sql(query, conn)
```

# Business Question 8

Which states generate the highest profit?

```python
query = """
SELECT
    state,
    ROUND(SUM(CAST(profit AS REAL)),2) AS total_profit
FROM superstore
GROUP BY state
ORDER BY total_profit DESC
LIMIT 10;
"""

pd.read_sql(query, conn)
```

# Business Question 9

Which products have sales above the average sales value?


```python
query = """
SELECT *
FROM superstore
WHERE CAST(sales AS REAL) >
(
    SELECT AVG(CAST(sales AS REAL))
    FROM superstore
)
LIMIT 10;
"""

pd.read_sql(query, conn)
```

# Business Question 10

Rank the top cities based on total sales.

```python
query = """
SELECT
    city,
    ROUND(SUM(CAST(sales AS REAL)),2) AS total_sales,
    RANK() OVER
    (
        ORDER BY SUM(CAST(sales AS REAL)) DESC
    ) AS sales_rank
FROM superstore
GROUP BY city
LIMIT 10;
"""

pd.read_sql(query, conn)
```
